### Strain vs Time and Stress vs Time Together

```python
import matplotlib.pyplot as plt
def plot_together(file_no):
    maindir=os.getcwd()
    strain_file_loc=maindir+'\\Data\\'
    stress_file_loc=maindir+'\\Data\\Tensile\\'
    strain_file_name=strain_file_loc+'Tensile_'+str(file_no)+'_ASCII_Vec\\*.txt'
    stress_file_name=stress_file_loc+'T'+str(file_no)+'.txt'
    time_file_name=strain_file_loc+'Tensile_'+str(file_no)+'_ASCII_Time\\B00001.txt'
    '''Import the Davis Strain ASCII files,Strain vs Time ASCII file and the Tensile data'''
    files=glob (strain_file_name);
    time_data=loadtxt(time_file_name[:])
    time_flatten=time_data.flatten()
    time=time_flatten[(len(time_flatten)/2):len(time_flatten)]-time_flatten[len(time_flatten)/2]
    stress_data=loadtxt(stress_file_name,skiprows=7,delimiter=',')[:,3]
    corrected_data=stress_data+(0-stress_data[-1])
    time_data=loadtxt(stress_file_name,skiprows=7,delimiter=',')[:,1]
    
    '''Blank lists to add the calculated strain values'''
    mean_strain_list=[]
    n_list=[]
    n1_list=[]
    strain_list=[]
    
    '''Going into each ASCII file and calcualting the average strain %'''
    for file in files:
        Vy_list=loadtxt(file)[:,3]
        plot_data=[]
        n_list=[]
        n1_list=[]
        for n,n1 in enumerate(Vy_list):
            n_list.append(n)
            n1_list.append(n1)
        for pos,vec in zip(n_list,n1_list):
            if abs(vec)>0.001:        
                strain=abs(((Vy_list[pos+22]-vec))/1.466) 
                '''The line above will change based on the interrogation window size which is 128 in this case'''                                             
                '''if function to get rid of the zero values'''
                if strain<0.1:
                    plot_data.append(strain)
        '''Cropping the top and the bottom as these cause noisy data towards higher strains'''
        strain_val=(average(plot_data[15:-10]))
        mean_strain_list.append(strain_val)
        
    stress_data=loadtxt(stress_file_name,skiprows=7,delimiter=',')[:,3]
    '''Correcting the stress data for the applied pre-load'''
    corrected_data=stress_data+(0-stress_data[-1])
    time_data=loadtxt(stress_file_name,skiprows=7,delimiter=',')[:,1]
    
    '''plotting stress vs time and strain vs time on the same graph'''
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot(time, mean_strain_list, '-', label = 'Strain')
    ax2 = ax.twinx()
    ax2.plot(time_data, corrected_data, '-r', label = 'Stress')
    ax.legend(loc=0)
    ax.legend(loc=1)
    ax.grid()
    ax.set_xlabel("Time (s)")
    ax.set_ylabel("True Strain")
    ax2.set_ylabel("Stress(MPa)")
    plt.show()
```

### Plotting the derivation to find the the turning point of the curve

```python
def plot_gradient(file_no):
    maindir=os.getcwd()
    strain_file_loc=maindir+'\\Data\\'
    stress_file_loc=maindir+'\\Data\\Tensile\\'
    strain_file_name=strain_file_loc+'Tensile_'+str(file_no)+'_ASCII_Vec\\*.txt'
    stress_file_name=stress_file_loc+'T'+str(file_no)+'.txt'
    time_file_name=strain_file_loc+'Tensile_'+str(file_no)+'_ASCII_Time\\B00001.txt'
    files=glob (strain_file_name);
    time_data=loadtxt(time_file_name[:])
    time_flatten=time_data.flatten()
    time=time_flatten[(len(time_flatten)/2):len(time_flatten)]-time_flatten[len(time_flatten)/2]
    mean_strain_list=[]
    n_list=[]
    n1_list=[]
    strain_list=[]
    strain_diff_list=[]
    stress_diff_list=[]
    mid_list=[]
    mid_time_list=[]
    for file in files:
        Vy_list=loadtxt(file)[:,3]
        plot_data=[]
        n_list=[]
        n1_list=[]
        for n,n1 in enumerate(Vy_list):
            n_list.append(n)
            n1_list.append(n1)
        for pos,vec in zip(n_list,n1_list):
            if abs(vec)>0.001:        
                strain=abs(((Vy_list[pos+22]-vec))/1.466)
                if strain<0.1:
                    plot_data.append(strain)
        strain_val=(average(plot_data[15:-10]))
        mean_strain_list.append(strain_val)
    stress_data=loadtxt(stress_file_name,skiprows=7,delimiter=',')[:,3]
    corrected_data=stress_data+(0-stress_data[-1])
    time_data=loadtxt(stress_file_name,skiprows=7,delimiter=',')[:,1]
    '''Calculating the difference for every 10 points. I don't think this method is very accurate'''
    
    for pos in linspace(0,len(corrected_data)-9,num=len(corrected_data)-9):
        difference=corrected_data[int(pos)+8]-corrected_data[int(pos)]
        stress_diff_list.append(difference)
    '''Calculating the difference for every 2 points. I don't think this method is very accurate'''
    for pos in linspace(0,len(mean_strain_list)-2,num=len(mean_strain_list)-2):
        difference=mean_strain_list[int(pos)+1]-mean_strain_list[int(pos)]
        strain_diff_list.append(difference)
    '''Calculating the median for 5 points in the difference in order to smooth out the curve'''
    cropped_time=time[:-2]
    for pos in linspace(0,len(strain_diff_list)-6,num=len(strain_diff_list)-6):
        mid=median(strain_diff_list[int(pos):int(pos)+5])
        mid_time=median(cropped_time[int(pos):int(pos)+5])
        mid_list.append(mid)
        mid_time_list.append(mid_time)
        
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot(mid_time_list,mid_list, '-', label = 'Strain_Diff')
    ax2 = ax.twinx()
    ax2.plot(time_data[:-9], stress_diff_list, '-r', label = 'Stress_Diff')
    ax.legend(loc=0)
    ax.legend(loc=1)
    ax.grid()
    ax.set_xlabel("Time (s)")
    ax.set_ylabel("Strain_Diff")
    ax2.set_ylabel("Stress_Diff")
    plt.show()
```

### Plotting the Stress Strain Curves

```python
def stress_strain_curve(x1,x2,x3,x4,x5):
    maindir=os.getcwd()
    strain_file_loc=maindir+'\\Data\\'
    stress_file_loc=maindir+'\\Data\\Tensile\\'
    fig = plt.figure()
    ax = fig.add_subplot(111)
    for file_no,marker in zip([x1,x2,x3,x4,x5],[".","o","s","*","_"]):
        strain_file_name=strain_file_loc+'Tensile_'+str(file_no)+'_ASCII_Vec\\*.txt'
        stress_file_name=stress_file_loc+'T'+str(file_no)+'.txt'
        time_file_name=strain_file_loc+'Tensile_'+str(file_no)+'_ASCII_Time\\B00001.txt'
        files=glob (strain_file_name);
        time_data=loadtxt(time_file_name[:])
        time_flatten=time_data.flatten()
        time=time_flatten[(len(time_flatten)/2):len(time_flatten)]-time_flatten[len(time_flatten)/2]
        mean_strain_list=[]
        n_list=[]
        n1_list=[]
        strain_list=[]
        strain_diff_list=[]
        stress_diff_list=[]
        mid_list=[]
        for file in files:
            Vy_list=loadtxt(file)[:,3]
            plot_data=[]
            n_list=[]
            n1_list=[]
            for n,n1 in enumerate(Vy_list):
                n_list.append(n)
                n1_list.append(n1)
            for pos,vec in zip(n_list,n1_list):
                if abs(vec)>0.001:        
                    strain=abs(((Vy_list[pos+22]-vec))/1.466)
                    if strain<0.1:
                        plot_data.append(strain)
            strain_val=(average(plot_data[30:-10]))
            mean_strain_list.append(strain_val)
        stress_data=loadtxt(stress_file_name,skiprows=7,delimiter=',')[:,3]
        corrected_data=stress_data+(0-stress_data[-1])
        time_data=loadtxt(stress_file_name,skiprows=7,delimiter=',')[:,1]
        '''Corrected stress data is interpolated against time'''
        zi=interp1d(time_data,corrected_data)
        '''The interpolated stress data values are matched against the time data for strain values'''
        yi=zi(time[2:])
        ax.plot(mean_strain_list[2:],yi, marker, label = 'Tensile'+str(file_no))
        ax.legend(loc=0)
        ax.grid()
        ax.set_xlabel("True Strain")
        ax.set_ylabel("True Stress(MPa)")
        plt.hold(True)
    plt.show()
```
