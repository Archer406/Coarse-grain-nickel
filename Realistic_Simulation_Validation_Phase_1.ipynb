# Introduction

Initially this report was started to investigate the effect of the element size on the tensile properties of the simulated gauge as very large values were observed. However during the writing of this report it became clear that these large values in tensile stresses were actually scaled up by the volume of the mesh. Hence the for a mesh with a volume of 7776 the obtained stress values have to be divided by 7776. 

Hence now Phase 1 serves to show if the simulation accurately predicts actual tensile results using the RS5_fmt material file. Phase 2 will involve observing the internal strain of the simulated gauge in order to see if it matches the experimentally observed DIC results.Phase 3 will involve carrying out the simulations on other tensile samples.

# Conclusion

Changing the element length scales the tensile value by the same factor that the volume is scaled up. Hence in single crystal simulations there is no effect seen.

However in polycrystalline simulation of the tensile gauge the volume used does have an effect on the calculated yield stress for the simulated gauge. When the element length is 1 in all directions and the size is not proportional to the number of elements the yield point is higher. The transistion from elastic to plastic is also broader when the mesh volume is one. However there is no difference in the simulated curves when the mesh volume is increased from 0.28 to 7776. This means that the tensile curve shape will remain consistent as long as the lengths of the elements remain proportional to the number of elements used. Both of these show a transistion from elastic to plastic that is similar to what is observed experimentally. The yield points are clearly lower than what is observed experimentally but this can be altered by increasing the crss value to between 400 and 500 (current crss value used is 350).

#### Collecting Stress Data from Tensile Rig

```python
def read_stress(x):
    maindir=os.getcwd()
    stress_file_loc=maindir+'/Data/Tensile/'
    for file_no in  [x]:#loading data from file
        stress_file_name=stress_file_loc+'T'+str(file_no)+'.txt'
    
    stress_data=loadtxt(stress_file_name,skiprows=7,delimiter=',')[:,3]
    corrected_data=stress_data+(0-stress_data[-1])#making correction as force is not zero in the beginning
    time_data=loadtxt(stress_file_name,skiprows=7,delimiter=',')[:,1]#time data from the same stress file
    return zip(corrected_data,time_data)
```

#### Collecting Time data from DIC data

```python
def strain_time(x):
    maindir=os.getcwd()
    strain_file_loc=maindir+'/Data/'
    for file_no in [x]:
        time_file_name=strain_file_loc+'Tensile_'+str(file_no)+'_ASCII_Time/B00001.txt'
        time_data=loadtxt(time_file_name[:])
        time_flatten=time_data.flatten()
        time=time_flatten[(len(time_flatten)/2):len(time_flatten)]-time_flatten[len(time_flatten)/2]
        time_file=strain_file_loc+'Tensile_'+str(file_no)+'_ASCII_Time/Time'+str(file_no)+'.txt'
        f=open(time_file,'w')
        f.write(time)
        f.close()
        corrected_time=time
    return corrected_time
```

#### Interpolating strain data with time data from DIC

```python
def interpolate(tensile_no):
    time=strain_time(tensile_no)
    mean_strain_list=read_strain(tensile_no)
    zi=interp1d(time[2:],mean_strain_list[2:])
    #this is done because generally the tensile rig data is 2 seconds behind the DIC data
    return zi
```

#### Interpolating the strain values with time data from tensile rig and calculating true stress and true strain

```python
def true_stress_strain(tensile_no):
    stress_time_data=read_stress(tensile_no)
    stress_list=[]
    time_data=[]
    zi=interpolate(tensile_no)

    for stress,time in stress_time_data:
        stress_list.append(stress)
        time_data.append(time)

    yi=zi(time_data[:-10])
    true_stress=[]

    for true_s,true_e in zip(stress_list[:len(yi)],yi):
        true_s=true_s*(1+true_e)
        true_stress.append(true_s)
    return zip(yi,true_stress)
```

#### Plotting True stress/strain graph

```python
def plot_graph(tensile_no):
    plot_values=true_stress_strain(tensile_no)
    true_strain=[]
    true_stress=[]
    for eyy,syy in plot_values:
        true_strain.append(eyy),true_stress.append(syy)        
    ax.plot(true_strain,true_stress,'-r',label = 'Tensile '+str(tensile_no)+'_Experimental')
    ax.grid()
    ax.legend(loc=0)
    ax.set_xlabel("True Strain")
    ax.set_ylabel("True Stress(MPa)")
    ax.hold(True)
```
